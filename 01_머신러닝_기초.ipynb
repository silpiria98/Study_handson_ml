## __1. 머신러닝이란?__
---

___Arthur Samuel, 1959___  
> '머신러닝은 명시적인 프로그래밍 없이 컴퓨터가 학습하는 능력을 갖추게 하는 연구 분야이다.' 
 

머신러닝은 컴퓨터에게 <b>데이터와 알고리즘을 주어 문제를 해결하는 방안을 도출하는 것</b>입니다.  
컴퓨터는 주어진 데이터를 알고리즘을 통해 학습하여 새로운 데이터에 대한 답안을 돌려줍니다.

#### __데이터 마이닝__

데이터 마이닝이란 <b>머신러닝이 학습한 결과를 확인하여 새로운 사실을 알아내는 것</b>입니다.  
주로 대규모 데이터베이스에서 새로운 추세나 연관관계 등을 발견하게 됩니다.

## __2. 머신러닝의 종류__  
---

### __2.1 학습하는 동안의 감독 형태나 정보량의 차이에 따른 분류__  


#### __지도 학습__
지도학습은 <b>데이터</b>와 <b>정답(레이블)</b>을 같이 주어 학습시키는 방법을 말합니다.
> <b>분류</b>: 데이터를 정해진 집단으로 나누는 작업  
<b>예측</b>: 다른 데이터를 바탕으로 원하는 값을 도출

#### __비지도 학습__
비지도학습은 <b>데이터만</b>을 주어 학습시키는 방법입니다.  
> <b>군집화</b>: 사전에 정해져 있지 않은 집단으로 분류  
<b>시각화</b>: 데이터를 그림으로 표현하여 사람의 눈으로 보기 쉽게 변환  
<b>차원축소</b>: 중요한 정보를 유지한 상태로 데이터의 양을 축소  
<b>이상치 탐지</b>: 데이터 내에 있는 이상한(확률이 적은) 데이터를 탐지  
<b>특이치 탐지</b>: 학습 데이터에 없는 새로운 데이터를 탐지  
<b>연관규칙 학습</b>: 데이터 간의 연관성 도출

#### __준지도 학습__
준지도학습은 <b>일부의 레이블</b>을 활용하는 학습 방법입니다.  

비용 등의 문제로 데이터에 일부 레이블이 없는 경우에 사용되며,  
보통 지도학습과 비지도학습의 병행으로 이루어집니다.
> <b>심층 신뢰 신경망, DBN</b>: 여러 겹의 <b>제한된 볼츠만 머신, RBM</b>으로 비지도학습을 한 후, 지도학습을 수행

#### __강화 학습__
강화강습은 여러 선택지 중에 <b>최선의 선택지</b>를 선택하게 학습시킵니다.  

보행로봇이나, 바둑, 체스 등의 게임AI 등에 사용합니다.

### __2.2 점진적 학습 가능 여부에 따른 분류__

#### __배치 학습, 오프라인 학습__
배치학습은 모든 데이터를 <b>한 번에 학습</b>하는 방법입니다.  

시간과 자원이 많이 필요하여 오프라인에서 학습을 하고, 제품 시스템에서는 적용만 하게됩니다.  
따라서 잦은 없데이트가 필요한 경우, 데이터 양이 너무 많은 경우, 컴퓨팅 성능이 한정된 경우 사용을 지양하는 것이 바람직합니다.

#### __온라인 학습__
온라인 학습은 데이터를 <b>순차적</b>으로 하나 또는 미니배치라는 단위로 처리합니다.  

배치학습의 단점을 보완할 수 있지만, 학습이 점진적으로 이루어지기 때문에 
나쁜 데이터가 들어올 경우 모델의 성능이 실시간으로 저하될 수 있습니다.  
또한 실시간으로 모델이 변화하되 예전 데이터가 무용지물이 될 수 있으므로, 적절한 학습률을 찾는 것이 중요합니다.


### __2.3 어떻게 모델을 일반화 시키는지에 따른 분류__

#### __사례기반 학습__
사례기반학습은 데이터를 기억하고, 새로운 데이터에 대한 <b>유사도</b>를 비교하여 판한합니다.

#### __모델기반 학습__
모델기반학습은 데이터로부터 <b>모델</b>을 만듭니다.  

모델은 최적의 결과를 내기 위한 <b>파라미터</b>를 찾는 과정입니다.  
일반적으로 모델이 얼마나 나쁜지를 나타내는 <b>비용함수(cost function)</b>을 정의하여 이를 최소화 합니다.  

## __3. 좋은 머신러닝의 조건__
---

### __좋은 데이터__
머신러닝은 데이터를 통하여 결과를 만들어가기 때문에 처음부터 데이터가 좋지 않으면 그 이후의 노력들도 무의미해 질 수 있습니다. 

좋은 데이터란 우선 데이터의 양이 충분해야 합니다. 또한 그 데이터들이 전체를 대표할 수 있어야 합니다. 에러, 이상치 등의 잡음도 없으면 좋겠죠. 가끔은 데이터의 멀쩡한 특성 자체가 머신러닝에 방해가 되기도 합니다. 그러한 특성들은 잡음과 마찬가지로 제거해 주는 것이 좋습니다.

### __나쁜 알고리즘__
<b>과대적합</b>은 모델이 훈련데이터에만 적합하고, 새로운 데이터에 대해서는 성능이 좋지 않은 경우를 말합니다.  
일반적으로 모델이 너무 복잡하거나, 데이터의 양이 부족하거나, 잡음이 너무 많은 경우에 나타나게 됩니다. 때문에 <b>규제</b>를 통하여 모델을 제약하거나 <b>차원축소</b>를 통해 특성을 줄여 모델의 복잡도를 줄이고, <b>전처리</b>를 통해 잡음을 제거하여 개선할 수 있습니다.  

<b>과소적합</b>은 반대로 모델이 너무 단순하여 훈련데이터 조차도 학습하지 못하는 경우입니다.  
과대적합과 반대로 특성을 늘리거나 규제를 약화하여 모델의 복잡도를 높여 개선할 수 있습니다.
